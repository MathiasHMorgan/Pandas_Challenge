```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Isolate Columns

Players = purchase_data.loc[:, ["Gender", "SN", "Age"]]
Players.head()
```

```python
#Remove duplicates and null values

Players = Players.drop_duplicates()
Players = Players.dropna(how='any')
```

```python
#Find player Count

Players_Count = Players.count()[0]
Players_Count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calcultions list

Unique_Items = len(purchase_data["Item ID"].unique())
Average_Price = purchase_data["Price"].mean()
Number_Purchase = purchase_data["Purchase ID"].count()
Total_Revenue = purchase_data["Price"].sum()
```

```python
#Summary Table

Purchase_Table = pd.DataFrame({"Number of Unique Items": [Unique_Items],
                               "Average Price": [Average_Price],
                               "Number of Purchases": [Number_Purchase],
                               "Total Revenue": [Total_Revenue]})

Purchase_Table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Establish Total Gender Count

Total_Genders = Players['Gender'].value_counts()
Total_Genders
```

```python
#Find Gender Percentage

Gender_Percentage = (Total_Genders/Players_Count)*100
Gender_Percentage
```

```python
#Summary Table

Final_Gender_Table = pd.DataFrame({"Total Count": Total_Genders, "Percentage of Players": Gender_Percentage})

Final_Gender_Table
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Total Gender Count

Total_Gender_Purchase_Count = purchase_data['Gender'].value_counts()
Total_Gender_Purchase_Count
```

```python
#Average Purchase Price Per Gender

Gender_Average_Purchase_Price = purchase_data.groupby(["Gender"])["Price"].mean()
Gender_Average_Purchase_Price
```

```python
#Average Purchase Price Per PErson

Average_Purchase_PerPerson = (Gender_Sum_Purchase_Value)/(Total_Gender_Purchase_Count)
Average_Purchase_PerPerson
```

```python
#Summary Table

Gender_Analysis_Table = pd.DataFrame({"Purchase Count": Total_Gender_Purchase_Count,
                               "Average Purchase Price": Gender_Average_Purchase_Price,
                               "Total Purchase Value": Gender_Sum_Purchase_Value,
                               "Avg Total Purchase Per Person": Average_Purchase_PerPerson})

Gender_Analysis_Table
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Visualise Data

purchase_data.head()
```

```python
#Bins and Labels

bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
#Isolate Age Range

purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
```

```python
#Calculations list

#Purchase Count
age_purchase_count = purchase_data.groupby(["Age Range"]).count()["Purchase ID"]

#Average Purchase Count
age_average_price = purchase_data.groupby(["Age Range"]).mean()["Price"]

#Total Value Purchased
age_total_purchase_value = purchase_data.groupby(["Age Range"]).sum()["Price"]

#Average Purcahse Price Per Person
average_purchase_per_person = age_total_purchase_value/age_purchase_count
```

```python
#Summary Table

Age_Table_Final = pd.DataFrame({"Purchase Count": age_purchase_count,
                               "Average Purchase Price": age_average_price,
                               "Total Purchase Value": age_total_purchase_value,
                               "Average Purchase Per Person": average_purchase_per_person})

Age_Table_Final
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Calculations List

SN_Count = purchase_data.groupby(["SN"]).count()["Purchase ID"]
SN_Average = purchase_data.groupby(["SN"]).mean()["Price"]
SN_Total = purchase_data.groupby(["SN"]).sum()["Price"]
```

```python
#Summary Table

Top_Spender = pd.DataFrame({"Purchase Count": SN_Count, 
                           "Average Purchase Price": SN_Average, 
                           "Total Purchase Value": SN_Total})

Top_Spender.head()
```

```python
#Sort Table

Arranged_Top_Spender = Top_Spender.sort_values("Total Purchase Value", ascending = False)

Arranged_Top_Spender.head()
```
